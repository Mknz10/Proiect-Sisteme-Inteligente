# Eliminare anomalii

Verificăm valorile null

```python
null = pd.DataFrame({'Null Values' : df.isna().sum(), 'Percentage Null Values' : (df.isna().sum()) / (df.shape[0]) * (100)})
null
```

Înlocuim null cu 0

```python
df.fillna(0, inplace = True)
```

Adulții, copii și bebelușii nu pot fi 0 deodată, deci eliminăm rândurile care au toate aceste valori 0

```python
filter = (df.children == 0) & (df.adults == 0) & (df.babies == 0)
df[filter]
```

```python
df = df[~filter]
df
```

# Histograme

```python
columns_to_plot = ['adr', 'lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies']
for col in columns_to_plot:
    plt.figure(figsize=(8, 5))
    df[col].hist(bins=50, edgecolor='black')
    plt.title(f'Distribuția {col}')
    plt.xlabel(col)
    plt.ylabel('Frecvența')
    plt.show()
```

# Gini index

Gini index pentru is_conceled

```python
canceled_counts = df['is_canceled'].value_counts(normalize=True)
p0 = canceled_counts[0]  
p1 = canceled_counts[1]

print(1 - (p0**2 + p1**2))
```

Gini index pentru tipurile de hotel

```python
hotel_counts = df['hotel'].value_counts(normalize=True)
p_resort = hotel_counts['Resort Hotel']
p_city = hotel_counts['City Hotel']

print(1 - (p_resort**2 + p_city**2))
```

# Entropia

```python
def entropy(probs):
    return -sum(p * np.log2(p) for p in probs if p > 0)

canceled_counts = df['is_canceled'].value_counts(normalize=True)
print(entropy(canceled_counts))

hotel_counts = df['hotel'].value_counts(normalize=True)
print(entropy(hotel_counts))
```

Coloane cu valori nenumerice:

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'O']
cat_df = df[cat_cols]
cat_df.head() 
```

Schimbam data aaaa-ll-zz in 3 coloane

```python
cat_df['reservation_status_date'] = pd.to_datetime(cat_df['reservation_status_date'])

cat_df['year'] = cat_df['reservation_status_date'].dt.year
cat_df['month'] = cat_df['reservation_status_date'].dt.month
cat_df['day'] = cat_df['reservation_status_date'].dt.day

cat_df.drop(['reservation_status_date','arrival_date_month'] , axis = 1, inplace = True)
```

```python
cat_df.head()
```

```python
for col in cat_df.columns:
    print(f"{col}: \n{cat_df[col].unique()}\n")
```

```python
cat_df['hotel'] = cat_df['hotel'].map({'Resort Hotel' : 0, 'City Hotel' : 1})

cat_df['meal'] = cat_df['meal'].map({'BB' : 0, 'FB': 1, 'HB': 2, 'SC': 3, 'Undefined': 4})

cat_df['market_segment'] = cat_df['market_segment'].map({'Direct': 0, 'Corporate': 1, 'Online TA': 2, 'Offline TA/TO': 3,'Complementary': 4, 'Groups': 5, 'Undefined': 6, 'Aviation': 7})

cat_df['distribution_channel'] = cat_df['distribution_channel'].map({'Direct': 0, 'Corporate': 1, 'TA/TO': 2, 'Undefined': 3,  'GDS': 4})

cat_df['reserved_room_type'] = cat_df['reserved_room_type'].map({'C': 0, 'A': 1, 'D': 2, 'E': 3, 'G': 4, 'F': 5, 'H': 6, 'L': 7, 'B': 8})

cat_df['deposit_type'] = cat_df['deposit_type'].map({'No Deposit': 0, 'Refundable': 1, 'Non Refund': 3})

cat_df['customer_type'] = cat_df['customer_type'].map({'Transient': 0, 'Contract': 1, 'Transient-Party': 2, 'Group': 3})
```
